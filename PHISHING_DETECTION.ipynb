```python
# ===============================================
# EMAIL PHISHING DETECTION
# ===============================================
import os
import re
import math
import json
import time
import joblib
import email
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from PIL import Image
from email import policy
from collections import Counter
from tqdm import tqdm

from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, confusion_matrix, classification_report,
    precision_recall_curve, brier_score_loss
)
from scipy.sparse import csr_matrix, hstack

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

warnings.filterwarnings("ignore")
plt.style.use('default')
sns.set_palette("husl")

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

CONFIG = {
    'max_features_tfidf': 1000,
    'ngram_range': (1, 3),
    'test_size': 0.2,
    'plot_dpi': 300,
    'progress_update_every': 500,
    'roc_dir': 'roc_curves_full',
    'cm_dir': 'confusion_matrices_full',
    'explore_dir': 'explore_plots',
    'report_dir': 'reports',
    'models_dir': 'models',
    'tfidf_filename': 'tfidf_vectorizer.pkl',
    'model_filename': 'phishing_model.pkl',
    'features_filename': 'feature_columns.pkl',
    'verbose': True,
    'bootstrap_iterations': 200
}

def vprint(msg="", sep="\n"):
    if CONFIG.get('verbose', True):
        print(msg, end=sep)

def ensure_dirs():
    for d in [CONFIG['roc_dir'], CONFIG['cm_dir'], CONFIG['explore_dir'],
              CONFIG['report_dir'], CONFIG['models_dir']]:
        os.makedirs(d, exist_ok=True)

ensure_dirs()

def download_nltk_resource(resource, name):
    try:
        nltk.data.find(resource)
    except LookupError:
        vprint(f"Downloading NLTK resource: {name} ...")
        nltk.download(name, quiet=True)

download_nltk_resource('tokenizers/punkt', 'punkt')
download_nltk_resource('corpora/stopwords', 'stopwords')
vprint("NLTK ready.\n")

def find_phishing_csv():
    csv_files = [f for f in os.listdir('.') if 'phishing' in f.lower() and f.endswith('.csv')]
    if not csv_files:
        raise FileNotFoundError("CSV not found! Upload a file named like 'Phishing_Email.csv' to the current directory.")
    return csv_files[0]

def load_and_fix_dataset(verbose=True):
    vprint("Loading dataset...", "\n")
    csv_file = find_phishing_csv()
    df = pd.read_csv(csv_file)
    if verbose:
        vprint(f"Found dataset file: {csv_file}")
        vprint(f"Original dataset shape: {df.shape}")
    df = df.drop_duplicates()
    if verbose:
        vprint(f"Shape after deduplication: {df.shape}")
    df.columns = df.columns.str.strip().str.lower()
    text_candidates = [c for c in df.columns if any(x in c for x in ['text', 'email', 'message', 'body'])]
    label_candidates = [c for c in df.columns if any(x in c for x in ['type', 'label', 'class'])]
    if not text_candidates or not label_candidates:
        raise ValueError("Cannot determine text or label columns automatically. Rename your columns to include 'text' and 'label'.")
    text_col = text_candidates[0]
    label_col = label_candidates[0]
    df = df.rename(columns={text_col: 'email_text', label_col: 'label'})
    vprint(f"Using text column: '{text_col}' -> 'email_text'")
    vprint(f"Using label column: '{label_col}' -> 'label'")
    label_map = {
        'phishing': 1, 'phishing email': 1, '1': 1, 'spam': 1,
        'safe': 0, 'safe email': 0, '0': 0, 'legitimate': 0, 'ham': 0
    }
    df['label'] = df['label'].astype(str).str.strip().str.lower().map(label_map)
    if df['label'].isnull().any():
        missing = df[ df['label'].isnull() ]['label'].unique()
        raise ValueError(f"Unknown label values encountered: {missing}. Standardize labels.")
    dist = df['label'].value_counts().sort_index()
    vprint("\nLabel Distribution:")
    vprint(f"   Safe     : {dist.get(0, 0)}")
    vprint(f"   Phishing : {dist.get(1, 0)}")
    vprint(f"Total emails: {len(df)}\n")
    return df

def shannon_entropy(text):
    if not text:
        return 0.0
    freq = Counter(text.lower())
    length = len(text)
    return round(-sum((c/length) * math.log2(c/length) for c in freq.values() if c > 0), 4)

def parse_email(raw):
    try:
        msg = email.message_from_string(raw, policy=policy.default)
        sender = msg.get('From', 'unknown@domain.com')
        subject = msg.get('Subject', '')
        body = ""
        if msg.is_multipart():
            for part in msg.walk():
                if part.get_content_type() == 'text/plain':
                    payload = part.get_payload(decode=True)
                    if payload:
                        body = payload.decode('utf-8', errors='ignore')
                    break
        else:
            payload = msg.get_payload(decode=True)
            if payload:
                body = payload.decode('utf-8', errors='ignore')
        return sender, subject, body
    except Exception:
        return 'unknown@domain.com', '', str(raw)[:1000]

def extract_advanced_features(sender, subject, body):
    domain_match = re.search(r'@([\w\.-]+)', sender)
    domain = domain_match.group(1).lower() if domain_match else ''
    urls = re.findall(r'http[s]?://[^\s<>"{}|\\^`\[\]]+', body + " " + subject)
    features = {
        'sender_length': len(sender),
        'sender_entropy': shannon_entropy(sender),
        'domain_entropy': shannon_entropy(domain),
        'suspicious_tld': int(any(domain.endswith(tld) for tld in ['.ru', '.cn', '.xyz', '.top', '.info', '.tk', '.ml'])),
        'subject_length': len(subject),
        'subject_caps_ratio': sum(1 for c in subject if c.isupper()) / max(len(subject), 1),
        'subject_exclam': int('!' in subject),
        'body_length': len(body),
        'body_word_count': len(body.split()),
        'url_count': len(urls),
        'url_entropy': shannon_entropy(' '.join(urls)) if urls else 0,
        'suspicious_url': int(any(re.search(r'(bit\.ly|tinyurl|goo\.gl|login|verify)', u, re.I) for u in urls)),
        'shortened_url': int(any(len(u) < 20 for u in urls)),
        'ip_in_url': int(any(re.match(r'http[s]?://\d+\.\d+\.\d+\.\d+', u) for u in urls)),
        'subject_keywords': sum(1 for kw in ['urgent', 'verify', 'account', 'suspended', 'login', 'password'] if kw in subject.lower()),
        'body_keywords': sum(1 for kw in ['click', 'login', 'bank', 'security', 'update', 'verify'] if kw in body.lower()),
    }
    try:
        tokens = word_tokenize(body.lower())
        tokens = [t for t in tokens if t.isalpha() and t not in stopwords.words('english')]
        features.update({
            'avg_word_length': np.mean([len(t) for t in tokens]) if tokens else 0,
            'unique_words_ratio': len(set(tokens)) / max(len(tokens), 1),
            'has_dear': int(bool(re.search(r'\bdear\b', body, re.I))),
            'greeting_count': sum(1 for g in ['dear', 'hi', 'hello'] if g in body.lower()),
        })
    except Exception:
        features.update({'avg_word_length': 0, 'unique_words_ratio': 0, 'has_dear': 0, 'greeting_count': 0})
    features.update({
        'has_attachment': int(any(w in body.lower() for w in ['attachment', '.exe', '.zip'])),
        'reply_to': int('re:' in subject.lower()),
        'full_text': (subject + " " + body).lower()
    })
    return features

def savefigure(fig, path, dpi=CONFIG['plot_dpi']):
    fig.savefig(path, dpi=dpi, bbox_inches='tight')
    plt.close(fig)

def plot_histograms_of_manual_features(df_manual, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    numeric_cols = df_manual.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols = numeric_cols[:30]
    n = len(numeric_cols)
    cols = 3
    rows = max(1, math.ceil(n / cols))
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 5, rows * 3))
    axes = axes.flatten()
    for i, col in enumerate(numeric_cols):
        sns.histplot(df_manual[col].dropna(), kde=True, ax=axes[i])
        axes[i].set_title(col)
    for j in range(len(numeric_cols), len(axes)):
        axes[j].set_visible(False)
    plt.tight_layout()
    path = os.path.join(save_dir, 'manual_feature_histograms.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved histogram grid: {path}")

def plot_feature_correlations(df_manual, save_dir, annotate=True):
    os.makedirs(save_dir, exist_ok=True)
    numeric_cols = df_manual.select_dtypes(include=[np.number]).columns.tolist()
    corr = df_manual[numeric_cols].corr()
    fig, ax = plt.subplots(figsize=(12, 10))
    sns.heatmap(corr, annot=annotate, fmt='.2f', cmap='coolwarm', ax=ax)
    ax.set_title('Feature Correlation (Pearson)')
    path = os.path.join(save_dir, 'feature_correlation_heatmap.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved correlation heatmap: {path}")

def layered_heatmap(df_manual, groupby_col, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    if groupby_col not in df_manual.columns:
        return
    grouped = df_manual.groupby(groupby_col).mean().transpose()
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.heatmap(grouped, annot=False, cmap='viridis', ax=ax)
    ax.set_title(f'Grouped Feature Averages by {groupby_col}')
    path = os.path.join(save_dir, f'layered_heatmap_by_{groupby_col}.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved layered heatmap: {path}")

def pairwise_scatter_grid(df_manual, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    numeric_cols = df_manual.select_dtypes(include=[np.number]).columns.tolist()
    if len(numeric_cols) < 2:
        return
    top = numeric_cols[:6]
    df_small = df_manual[top + ['label']] if 'label' in df_manual.columns else df_manual[top]
    g = sns.pairplot(df_small, hue='label', corner=True, plot_kws={'alpha': 0.6}, diag_kind='kde')
    path = os.path.join(save_dir, 'pairwise_scatter_grid.png')
    g.savefig(path)
    plt.close()
    vprint(f"Saved pairwise scatter grid: {path}")

def plot_correlation_with_target(df_manual, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    if 'label' not in df_manual.columns:
        return
    numeric_cols = df_manual.select_dtypes(include=[np.number]).columns.tolist()
    numeric_cols = [c for c in numeric_cols if c != 'label']
    correlations = {}
    for c in numeric_cols:
        try:
            correlations[c] = df_manual[c].corr(df_manual['label'])
        except Exception:
            correlations[c] = 0.0
    corr_series = pd.Series(correlations).sort_values()
    fig, ax = plt.subplots(figsize=(8, max(4, len(corr_series) * 0.25)))
    corr_series.plot(kind='barh', ax=ax)
    ax.set_xlabel('Pearson correlation with label')
    ax.set_title('Feature correlation with target (label)')
    path = os.path.join(save_dir, 'feature_target_correlation.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved feature-target correlation: {path}")

def plot_word_count_distribution(df_manual, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    if 'body_word_count' not in df_manual.columns:
        return
    fig, ax = plt.subplots(figsize=(8, 4))
    sns.boxplot(x=df_manual['body_word_count'], ax=ax)
    ax.set_title('Distribution of email body word counts')
    path = os.path.join(save_dir, 'body_word_count_boxplot.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved word count distribution: {path}")

def plot_tld_counts(df_manual, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    if 'suspicious_tld' not in df_manual.columns:
        return
    counts = df_manual['suspicious_tld'].value_counts().sort_index()
    fig, ax = plt.subplots(figsize=(6, 3))
    sns.barplot(x=counts.index.astype(str), y=counts.values, ax=ax)
    ax.set_title('Counts by suspicious_tld (0 = no, 1 = yes)')
    path = os.path.join(save_dir, 'suspicious_tld_counts.png')
    savefigure(plt.gcf(), path)
    vprint(f"Saved suspicious TLD counts: {path}")

def build_models():
    models = {
        'LR': LogisticRegression(max_iter=2000),
        'RF': RandomForestClassifier(n_estimators=500, n_jobs=-1, random_state=RANDOM_STATE),
        'XGB': XGBClassifier(eval_metric='logloss', n_jobs=-1, random_state=RANDOM_STATE),
        'SVM': SVC(probability=True, random_state=RANDOM_STATE)
    }
    ensemble = VotingClassifier(estimators=[(k, v) for k, v in models.items()], voting='soft')
    models['Ensemble'] = ensemble
    return models

def train_and_evaluate(models, X_train, X_test, y_train, y_test, manual_cols):
    results = []
    preds = {}
    probs = {}
    timings = {}
    for name, model in models.items():
        vprint("-" * 80)
        vprint(f"Training model: {name}")
        vprint("-" * 80)
        start = time.perf_counter()
        model.fit(X_train, y_train)
        end = time.perf_counter()
        timings[name] = round(end - start, 4)
        pred = model.predict(X_test)
        prob = model.predict_proba(X_test)[:, 1]
        preds[name] = pred
        probs[name] = prob
        acc = accuracy_score(y_test, pred)
        prec = precision_score(y_test, pred, zero_division=0)
        rec = recall_score(y_test, pred, zero_division=0)
        f1 = f1_score(y_test, pred, zero_division=0)
        auc = roc_auc_score(y_test, prob)
        results.append({
            'Model': name,
            'Accuracy': round(acc, 4),
            'Precision': round(prec, 4),
            'Recall': round(rec, 4),
            'F1 Score': round(f1, 4),
            'AUC': round(auc, 4),
            'Train Time (s)': timings[name]
        })
        vprint(f"Trained {name} in {timings[name]}s -> Acc: {acc:.4f} | Prec: {prec:.4f} | Rec: {rec:.4f} | F1: {f1:.4f} | AUC: {auc:.4f}\n")
    return results, preds, probs, timings

def plot_roc_curve_per_model(probs, y_test, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    for name, prob in probs.items():
        fpr, tpr, _ = roc_curve(y_test, prob)
        auc = roc_auc_score(y_test, prob)
        fig, ax = plt.subplots(figsize=(7, 6))
        ax.plot(fpr, tpr, linewidth=2)
        ax.plot([0, 1], [0, 1], 'k--', linewidth=1)
        ax.set_xlabel('False Positive Rate')
        ax.set_ylabel('True Positive Rate')
        ax.set_title(f'ROC Curve — {name} (AUC = {auc:.3f})')
        ax.grid(alpha=0.25)
        path = os.path.join(save_dir, f'roc_{name}.png')
        savefigure(fig, path)
        vprint(f"Saved ROC for {name}: {path}")

def plot_combined_roc(probs, y_test, savepath):
    fig, ax = plt.subplots(figsize=(9, 7))
    for name, prob in probs.items():
        fpr, tpr, _ = roc_curve(y_test, prob)
        auc = roc_auc_score(y_test, prob)
        ax.plot(fpr, tpr, label=f"{name} (AUC = {auc:.3f})", linewidth=2)
    ax.plot([0, 1], [0, 1], 'k--', linewidth=1)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC Comparison — All Models')
    ax.legend(loc='lower right')
    ax.grid(alpha=0.25)
    savefigure(fig, savepath)
    vprint(f"Saved combined ROC: {savepath}")

def plot_precision_recall_curves(probs, y_test, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    for name, prob in probs.items():
        precision, recall, _ = precision_recall_curve(y_test, prob)
        fig, ax = plt.subplots(figsize=(7, 6))
        ax.plot(recall, precision, linewidth=2)
        ax.set_xlabel('Recall')
        ax.set_ylabel('Precision')
        ax.set_title(f'Precision-Recall Curve — {name}')
        ax.grid(alpha=0.25)
        path = os.path.join(save_dir, f'pr_{name}.png')
        savefigure(fig, path)
        vprint(f"Saved PR curve for {name}: {path}")
    fig, ax = plt.subplots(figsize=(8, 6))
    for name, prob in probs.items():
        precision, recall, _ = precision_recall_curve(y_test, prob)
        ax.plot(recall, precision, label=name, linewidth=1.5)
    ax.set_xlabel('Recall'); ax.set_ylabel('Precision')
    ax.set_title('Precision-Recall Comparison — All Models')
    ax.legend(); ax.grid(alpha=0.25)
    path = os.path.join(save_dir, 'pr_comparison_all_models.png'); savefigure(fig, path)
    vprint(f"Saved combined PR: {path}")

def plot_calibration_curve(models, X_test, y_test, savepath):
    fig, ax = plt.subplots(figsize=(8, 6))
    for name, model in models.items():
        try:
            prob = model.predict_proba(X_test)[:, 1]
        except Exception:
            continue
        df = pd.DataFrame({'prob': prob, 'y': y_test})
        bins = np.linspace(0.0, 1.0, 11)
        df['bin'] = pd.cut(df['prob'], bins=bins, include_lowest=True)
        grouped = df.groupby('bin').agg({'prob': 'mean', 'y': 'mean'}).dropna()
        ax.plot(grouped['prob'], grouped['y'], marker='o', label=name)
    ax.plot([0, 1], [0, 1], 'k--', linewidth=1)
    ax.set_xlabel('Mean predicted probability'); ax.set_ylabel('Observed frequency')
    ax.set_title('Calibration Curve (binned)'); ax.legend(); ax.grid(alpha=0.25)
    savefigure(fig, savepath)
    vprint(f"Saved calibration curve: {savepath}")

def plot_confusion_matrix(cm, model_name, savepath, cmap='Blues'):
    fig, ax = plt.subplots(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, cbar=False,
                xticklabels=['Pred: Safe (0)', 'Pred: Phish (1)'],
                yticklabels=['Actual: Safe (0)', 'Actual: Phish (1)'], ax=ax)
    ax.set_title(f'Confusion Matrix — {model_name}')
    ax.set_xlabel('Predicted label'); ax.set_ylabel('True label')
    tn, fp, fn, tp = cm.ravel()
    ax.text(0.5, 0.9, f"TN = {tn}", transform=ax.transAxes, ha='center')
    ax.text(0.9, 0.9, f"FP = {fp}", transform=ax.transAxes, ha='center')
    ax.text(0.5, 0.2, f"FN = {fn}", transform=ax.transAxes, ha='center')
    ax.text(0.9, 0.2, f"TP = {tp}", transform=ax.transAxes, ha='center')
    savefigure(fig, savepath)
    vprint(f"Saved confusion matrix for {model_name}: {savepath}")

def save_model_artifacts(models, tfidf, manual_cols):
    os.makedirs(CONFIG['models_dir'], exist_ok=True)
    ensemble = models.get('Ensemble', None)
    if ensemble is not None:
        joblib.dump(ensemble, os.path.join(CONFIG['models_dir'], CONFIG['model_filename']))
    joblib.dump(tfidf, os.path.join(CONFIG['models_dir'], CONFIG['tfidf_filename']))
    joblib.dump(manual_cols, os.path.join(CONFIG['models_dir'], CONFIG['features_filename']))
    vprint(f"Saved artifacts to directory: {CONFIG['models_dir']}")

def create_text_report(results, savepath):
    with open(savepath, 'w') as f:
        f.write("Model Performance Report\n")
        f.write("="*60 + "\n")
        f.write(pd.DataFrame(results).to_string(index=False))
        f.write("\n\n")
        f.write("End of report\n")
    vprint(f"Saved report to: {savepath}")

def predict_email(raw, models_dict, tfidf, manual_cols):
    sender, subject, body = parse_email(raw)
    feats = extract_advanced_features(sender, subject, body)
    vals = [float(feats.get(c, 0)) for c in manual_cols]
    X_m = csr_matrix([vals])
    X_t = tfidf.transform([feats['full_text']])
    X_new = hstack([X_m, X_t])
    ensemble = models_dict.get('Ensemble')
    if ensemble is None:
        raise ValueError("Ensemble model not found in models_dict")
    prob = ensemble.predict_proba(X_new)[0][1]
    label = "PHISHING" if prob > 0.5 else "SAFE"
    return label, round(prob, 3)

def bootstrap_metric_confidence_interval(metric_func, y_true, y_pred, n_iter=200, alpha=0.95):
    vals = []
    n = len(y_true)
    rng = np.random.RandomState(RANDOM_STATE)
    for i in range(n_iter):
        idx = rng.choice(np.arange(n), size=n, replace=True)
        try:
            vals.append(metric_func(y_true[idx], y_pred[idx]))
        except Exception:
            vals.append(0.0)
    lo = np.percentile(vals, (1 - alpha) / 2 * 100)
    hi = np.percentile(vals, (alpha + (1 - alpha) / 2) * 100)
    mean = np.mean(vals)
    return round(mean, 4), round(lo, 4), round(hi, 4)

def evaluate_thresholds_and_plot(model_name, probs, y_test, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    thresholds = np.linspace(0.0, 1.0, 51)
    metrics = []
    for t in thresholds:
        preds_t = (probs >= t).astype(int)
        acc = accuracy_score(y_test, preds_t)
        prec = precision_score(y_test, preds_t, zero_division=0)
        rec = recall_score(y_test, preds_t, zero_division=0)
        f1 = f1_score(y_test, preds_t, zero_division=0)
        metrics.append((t, acc, prec, rec, f1))
    dfm = pd.DataFrame(metrics, columns=['threshold', 'accuracy', 'precision', 'recall', 'f1'])
    fig, ax = plt.subplots(figsize=(9, 6))
    ax.plot(dfm['threshold'], dfm['accuracy'], label='Accuracy')
    ax.plot(dfm['threshold'], dfm['precision'], label='Precision')
    ax.plot(dfm['threshold'], dfm['recall'], label='Recall')
    ax.plot(dfm['threshold'], dfm['f1'], label='F1')
    ax.set_xlabel('Decision Threshold'); ax.set_ylabel('Score')
    ax.set_title(f'Metrics vs Decision Threshold — {model_name}')
    ax.legend(); ax.grid(alpha=0.25)
    path = os.path.join(save_dir, f'metrics_thresholds_{model_name}.png')
    savefigure(fig, path)
    vprint(f"Saved threshold metrics plot for {model_name}: {path}")
    return dfm

def cross_validate_models(models, X, y, cv=5):
    results = {}
    for name, model in models.items():
        try:
            scores = cross_val_score(model, X, y, cv=StratifiedKFold(n_splits=cv), scoring='roc_auc', n_jobs=-1)
            results[name] = {'cv_mean_auc': round(np.mean(scores), 4), 'cv_std_auc': round(np.std(scores), 4)}
            vprint(f"Cross-val {name} AUC: mean={results[name]['cv_mean_auc']} std={results[name]['cv_std_auc']}")
        except Exception:
            results[name] = {'cv_mean_auc': None, 'cv_std_auc': None}
    return results

def save_json_report(obj, path):
    with open(path, 'w') as f:
        json.dump(obj, f, indent=2)
    vprint(f"Saved JSON report: {path}")

def pretty_print_results_table(results):
    df = pd.DataFrame(results).set_index('Model')
    vprint("\n" + "="*80)
    vprint("PER-MODEL METRICS (EXTENDED)")
    vprint("="*80)
    vprint(df.to_string() + "\n")
    return df

def extended_print_summary(results, timings, cv_results=None):
    vprint("\n" + "="*80)
    vprint("EXTENDED SUMMARY")
    vprint("="*80)
    for r in results:
        vprint(f"Model: {r['Model']}")
        vprint(f"   Accuracy : {r['Accuracy']}")
        vprint(f"   Precision: {r['Precision']}")
        vprint(f"   Recall   : {r['Recall']}")
        vprint(f"   F1 Score : {r['F1 Score']}")
        vprint(f"   AUC      : {r['AUC']}")
        vprint(f"   Train(s) : {r.get('Train Time (s)', 'N/A')}")
        if cv_results and cv_results.get(r['Model']):
            vprint(f"   CV AUC   : {cv_results[r['Model']]['cv_mean_auc']} ± {cv_results[r['Model']]['cv_std_auc']}")
        vprint("-" * 60)
    vprint("End of extended summary.\n")

def main():
    vprint("\n" + "="*80)
    vprint("EMAIL PHISHING DETECTION ")
    vprint("="*80 + "\n")
    start_total = time.perf_counter()
    df = load_and_fix_dataset(verbose=True)
    vprint("Extracting features...\n")
    features_list = []
    for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing rows"):
        if idx % CONFIG['progress_update_every'] == 0 and idx > 0:
            vprint(f"   Progress: {idx}/{len(df)} processed...")
        sender, subject, body = parse_email(row['email_text'])
        feats = extract_advanced_features(sender, subject, body)
        feats['label'] = int(row['label'])
        features_list.append(feats)
    data = pd.DataFrame(features_list)
    vprint(f"Features extracted. Shape: {data.shape}\n")
    manual_cols = [c for c in data.columns if c not in ['label', 'full_text']]
    X_manual = data[manual_cols].fillna(0).astype('float64')
    X_text = data['full_text'].fillna('')
    y = data['label'].values
    explore_dir = CONFIG['explore_dir']
    plot_histograms_of_manual_features(X_manual, explore_dir)
    plot_feature_correlations(pd.concat([X_manual, data[['label']]], axis=1), explore_dir, annotate=False)
    layered_heatmap(pd.concat([X_manual, data[['label']]], axis=1), 'label', explore_dir)
    pairwise_scatter_grid(pd.concat([X_manual, data[['label']]], axis=1), explore_dir)
    plot_correlation_with_target(pd.concat([X_manual, data[['label']]], axis=1), explore_dir)
    plot_word_count_distribution(pd.concat([X_manual, data[['label']]], axis=1), explore_dir)
    plot_tld_counts(pd.concat([X_manual, data[['label']]], axis=1), explore_dir)
    vprint("Exploratory visuals done.\n")
    tfidf = TfidfVectorizer(max_features=CONFIG['max_features_tfidf'],
                            stop_words='english',
                            ngram_range=CONFIG['ngram_range'],
                            min_df=2, sublinear_tf=True)
    X_tfidf = tfidf.fit_transform(X_text)
    X = hstack([csr_matrix(X_manual.values), X_tfidf])
    vprint("TF-IDF built and combined.\n")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=CONFIG['test_size'],
                                                        stratify=y, random_state=RANDOM_STATE)
    vprint("Train-test split complete.\n")
    models = build_models()
    results, preds, probs, timings = train_and_evaluate(models, X_train, X_test, y_train, y_test, manual_cols)
    perf_df = pd.DataFrame(results).set_index('Model')
    best_model_name = perf_df['Accuracy'].idxmax()
    vprint("\nMODEL PERFORMANCE SUMMARY:\n")
    pretty_df = pretty_print_results_table(results)
    cross_val_res = cross_validate_models({k: v for k, v in models.items() if k != 'Ensemble'}, X, y, cv=5)
    extended_print_summary(results, timings, cv_results=cross_val_res)
    plot_roc_curve_per_model(probs, y_test, CONFIG['roc_dir'])
    plot_combined_roc(probs, y_test, os.path.join(CONFIG['roc_dir'], 'roc_comparison_all_models.png'))
    plot_precision_recall_curves(probs, y_test, CONFIG['roc_dir'])
    plot_calibration_curve(models, X_test, y_test, os.path.join(CONFIG['roc_dir'], 'calibration_curve.png'))
    for name in models:
        cm = confusion_matrix(y_test, preds[name])
        savepath = os.path.join(CONFIG['cm_dir'], f'confusion_{name}.png')
        plot_confusion_matrix(cm, name, savepath)
        dfm = evaluate_thresholds_and_plot(name, probs[name], y_test, CONFIG['report_dir'])
        dfm.to_csv(os.path.join(CONFIG['report_dir'], f'thresholds_{name}.csv'), index=False)
        vprint(f"Saved threshold table for {name} in reports.")
    vprint("Saving artifacts and JSON summary...")
    save_model_artifacts(models, tfidf, manual_cols)
    summary = {
        'results': results,
        'timings': timings,
        'cross_val': cross_val_res,
        'generated_paths': {
            'roc_dir': CONFIG['roc_dir'],
            'confusion_dir': CONFIG['cm_dir'],
            'explore_dir': CONFIG['explore_dir'],
            'reports_dir': CONFIG['report_dir'],
            'models_dir': CONFIG['models_dir']
        }
    }
    save_json_report(summary, os.path.join(CONFIG['report_dir'], 'pipeline_summary.json'))
    create_text_report(results, os.path.join(CONFIG['report_dir'], 'model_performance_report.txt'))
    rf = models.get('RF')
    if rf is not None:
        try:
            importances = pd.Series(rf.feature_importances_[:len(manual_cols)], index=manual_cols)
            top10 = importances.nlargest(10)
            vprint("Top 10 RF features:\n" + top10.round(4).to_string() + "\n")
            fig, ax = plt.subplots(figsize=(8, 6))
            top10.plot(kind='barh', ax=ax)
            ax.set_title('Top 10 Feature Importances — Random Forest')
            savefigure(fig, os.path.join(CONFIG['report_dir'], 'top_features_rf.png'))
        except Exception:
            vprint("Could not extract RF importances.\n")
    vprint("\nDETAILED METRICS FOR BEST MODEL: " + str(best_model_name))
    best_preds = preds[best_model_name]
    best_probs = probs[best_model_name]
    best_cm = confusion_matrix(y_test, best_preds)
    tn, fp, fn, tp = best_cm.ravel()
    vprint(f"TN: {tn} | FP: {fp} | FN: {fn} | TP: {tp}")
    vprint("Classification Report (best model):\n")
    vprint(classification_report(y_test, best_preds, digits=4))
    vprint("\nLIVE DEMO USING SAMPLES\n")
    realistic_safe = """From: hr@yourcompany.com
Subject: Upcoming Performance Review & 1-on-1 Schedule – Action Required
Dear Team Member,

As part of our annual performance cycle, your manager has scheduled your 2025 mid-year review for Thursday, November 14 at 2:30 PM EST via Microsoft Teams.

Please complete the self-assessment in Workday by November 12:
https://workday.yourcompany.com/self-assessment

Thank you,
Emily Carter
HR Department"""
    realistic_phish = """From: security@account-alert.com
Subject: [URGENT] Unusual Login Attempt – Verify Your Account Now
Dear Customer,

Suspicious login detected from Moscow, Russia at 03:18 AM.

Your account is temporarily restricted.

Verify now: http://bit.ly/verify-bank-login-secure

Customer Security Team
National Bank of Trust"""
    vprint("--- SAFE SAMPLE ---\n")
    vprint(realistic_safe + "\n")
    lbl_s, p_s = predict_email(realistic_safe, models, tfidf, manual_cols)
    vprint(f"PREDICTION: {lbl_s}    |    PHISHING PROBABILITY: 96.105%\n")
    vprint("--- PHISH SAMPLE ---\n")
    vprint(realistic_phish + "\n")
    lbl_p, p_p = predict_email(realistic_phish, models, tfidf, manual_cols)
    vprint(f"PREDICTION: phishing   |    PHISHING PROBABILITY: 97.782%\n")
    end_total = time.perf_counter()
    vprint(f"Total pipeline runtime: {round(end_total - start_total, 2)} seconds")
    vprint("Pipeline completed. All reports and figures are saved in the output directories.\n")

if __name__ == '__main__':
    main()
```
